```python
import selenium
from selenium import webdriver as wd
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
from itertools import repeat
```

### 2. 크롬드라이버가 url 접속하도록 하기
https://blog.naver.com/song_sec/221752226329

```python
# 크롬드라이버 열기
driver = wd.Chrome() # 크롬드라이버 경로
driver.maximize_window() # 크롬창 크기 최대

# 드라이버가 해당 url 접속
url = 'https://www.melon.com/chart/index.htm' # 멜론차트 페이지
driver.get(url)
```

### 3. 크롬드라이버가 필터를 설정하도록 하기

```python
# 차트파인더 클릭
driver.find_element_by_xpath('//*[@id="gnb_menu"]/ul[1]/li[1]/div/div/button/span').click()

# 연대선택, 연도선택, 월선택, 장르선택

# 월간차트 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/h4[1]/a').click()
time.sleep(0.1)

# 연대선택 2010년 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[2]/span/label').click()
time.sleep(0.1)

# 연도선택 2018년 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[2]/span/label').click()
time.sleep(0.1)

# 월간선택 6월 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[3]/div[1]/ul/li[6]/span/label').click()
time.sleep(0.1)

# 주간선택 1주차 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[4]/div[1]/ul/li[1]/span/label').click()
time.sleep(0.1)

# 장르선택 종합 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[1]/span/label').click()
time.sleep(0.1)

# 검색버튼 클릭
driver.find_element_by_xpath('//*[@id="d_srch_form"]/div[2]/button/span/span').click()
```

### 4. html 정보 가져오기

```python
html = driver.page_source # 드라이버 현재 페이지의 html 정보 가져오기 
                            # cf) requests.get(url)
soup = BeautifulSoup(html, 'lxml')
```

### 5. 곡명 가져오기

```python
soup.find_all('div', attrs={'class': 'ellipsis rank01'})
```

```python
[title.find('a').get_text() for title in soup.find_all('div', attrs={'class': 'ellipsis rank01'})]
```

### 6. 가수명 가져오기

```python
soup.find_all('span', attrs={'class':'checkEllipsis'})
```

```python
[ singer.get_text() for singer in soup.find_all('span', attrs={'class':'checkEllipsis'}) ]
```

### 7. 순위 가져오기

```python
song = [title.find('a').get_text() for title in soup.find_all('div', attrs={'class': 'ellipsis rank01'})]
rank = []
for i in range(len(song)):
    rank.append(i+1)
rank
```
