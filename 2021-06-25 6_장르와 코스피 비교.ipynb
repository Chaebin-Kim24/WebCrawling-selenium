### 환경설정

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import sqlite3
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 글꼴 설정
conn = sqlite3.connect('song.db') # DB 경로
```

### SQLite에서 데이터 다운

```python
con = sqlite3.connect("db/new_song.db")
df_song_all_week = pd.read_sql("SELECT * FROM All_TOP100_week_NOST", con)
con.close()

df_song_all_week
```

```python
con = sqlite3.connect("db/new_song.db")
df_song_all_info = pd.read_sql("SELECT * FROM song_rank_info_NOST", con)
con.close()

df_song_all_info
```

```python
df_song_all = df_song_all_info[df_song_all_info.columns]
df_song_all['week'] = df_song_all_week['week']
df_song_all
```

### DataFrame 수정

```python
df_date_info = pd.read_csv('csv/date_info/date_info.csv')
df_date_info
```

```python
balad_list = []
dance_list = []
rap_hiphop_list = []

for date in df_date_info['date'] :
    balad = len(df_song_all[(df_song_all['week'] == date) & (df_song_all['genre'] == '발라드')])
    dance = len(df_song_all[(df_song_all['week'] == date) & (df_song_all['genre'] == '댄스')])
    rap_hiphop = len(df_song_all[(df_song_all['week'] == date) & (df_song_all['genre'] == '랩/힙합')])
    
    balad_list.append(balad)
    dance_list.append(dance)
    rap_hiphop_list.append(rap_hiphop)
    
df_date_genre = df_date_info[['date']]
df_date_genre['balad'] = balad_list
df_date_genre['dance'] = dance_list
df_date_genre['rap/hiphop'] = rap_hiphop_list

df_date_genre[:30]
```

### 그래프 그리기

```python
plt.figure(figsize=(15,6)) # 그림 크기
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.title("발라드 선호도")
g1_1=sns.scatterplot(data=df_date_genre, x='date', y="balad", legend=False )
g1_1.set(xticklabels=[])  # remove the tick labels  # add a title
g1_2=sns.lineplot(data=df_date_genre, x='date', y="balad")
```

```python
plt.figure(figsize=(15,6)) # 그림 크기
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.title("댄스 선호도")
g1_1=sns.scatterplot(data=df_date_genre, x='date', y="dance", legend=False )
g1_1.set(xticklabels=[])  # remove the tick labels  # add a title
g1_2=sns.lineplot(data=df_date_genre, x='date', y="dance")
```

```python
plt.figure(figsize=(15,6)) # 그림 크기
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.title("랩/힙합 선호도")
g1_1=sns.scatterplot(data=df_date_genre, x='date', y="rap/hiphop", legend=False )
g1_1.set(xticklabels=[])  # remove the tick labels  # add a title
g1_2=sns.lineplot(data=df_date_genre, x='date', y="rap/hiphop")
```
